```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

path=input(r"Enter file path: ")
if path.endswith('.csv'):
    df = pd.read_csv(path,encoding='unicode_escape')
elif path.endswith(('.xls', '.xlsx')):
    df = pd.read_excel(path)
else:
    raise ValueError("Unsupported file format")
    
df=df.drop_duplicates()
df.fillna(df.select_dtypes(include=['number']).mean(),inplace=True)
df.columns=df.columns.str.upper()

def empty():
    print (df.isnull().sum())
def shape():
    print (df.shape)
def dtypes():
    print (df.dtypes)
def columns():
    print (df.columns)
def head():    
    print (df.head(10))
def tail():
    print (df.tail(10))
def mean():
    print (df.select_dtypes(include='number').columns)
    ip=input("\nEnter column name to find its mean(Please enter correct spelling):")
    ip=ip.upper()
    print ("\nThe mean of",ip,"is:",df[ip].mean())
def median():
    print (df.select_dtypes(include='number').columns)
    ip=input("\nEnter column name to find its median(Please enter correct spelling):")
    ip=ip.upper()
    print ("\nThe median of",ip,"is:",df[ip].median())
def mode():
    print (df.select_dtypes(include='number').columns)
    ip=input("\nEnter column name to find its mode(Please enter correct spelling):")
    ip=ip.upper()
    print ("\nThe mode of",ip,"is:",df[ip].mode())
def std():
    print (df.select_dtypes(include='number').columns)
    ip=input("\nEnter column name to find its standard deviation(Please enter correct spelling):")
    ip=ip.upper()
    print ("\nThe standard deviation of",ip,"is:",df[ip].std())
def var():
    print (df.select_dtypes(include='number').columns)
    ip=input("\nEnter column name to find its variance(Please enter correct spelling):")
    ip=ip.upper()
    print ("\nThe variance of",ip,"is:",df[ip].var())
def des():
    print(df.describe())
    
def hist():
    print ("\nHistogram performance of one column")
    print (df.columns)
    ip=input("\nEnter column name to display its Histogram(Please enter correct spelling):")
    ip=ip.upper()
    plt.hist(df[ip],bins=20,color="red",edgecolor="black")
    plt.title(ip)
    plt.show()
def bar():
    print ("\nBar graph to show comparision of 2 columns")
    print (df.columns)
    ip1=input("\nEnter x axis column name for bar graph(Please enter correct spelling):")
    ip1=ip1.upper()
    ip2=input("\nEnter y axis column name for bar graph(Please enter correct spelling):")
    ip2=ip2.upper()
    plt.barh(df[ip1],df[ip2],color="red",edgecolor="black")
    plt.title("Bar Graph")
    plt.xlabel(ip2)
    plt.ylabel(ip1)
    plt.show()
def scatter():
    print ("\nScatter plot for checking Correlation between 2 columns")
    print (df.columns)
    ip1=input("\nEnter x axis column name for scatter plot(Please enter correct spelling):")
    ip1=ip1.upper()
    ip2=input("\nEnter y axis column name for scatter plot(Please enter correct spelling):")
    ip2=ip2.upper()
    plt.scatter(df[ip1],df[ip2],color="red")
    plt.title("Scatter Plot")
    plt.xlabel(ip1)
    plt.ylabel(ip2)
    plt.show()
def line():    
    print ("\nLine graph to compare 2 columns")
    print (df.columns)
    ip1=input("\nEnter column name for x axis:(Please enter correct spelling)")
    ip1=ip1.upper()
    ip2=input("\nEnter column name for y axis:(Please enter correct spelling)")
    ip2=ip2.upper()
    plt.plot(df[ip1],df[ip2],marker="o",linestyle="--")
    plt.title("Line Graph")
    plt.xlabel(ip1)
    plt.ylabel(ip2)
    plt.show()
def heatmap(): 
    print ("\nHeatmap for checking Correlation between columns")
    plt.figure(figsize=(10, 8))
    sns.heatmap(df.corr(numeric_only=True), annot=True, cmap='autumn', fmt=".2f")
    plt.title("Correlation Heatmap")
    plt.show()
def pie():
    print ("\nPie chart to show comparision of columns")
    df_columns=df.select_dtypes(include='number').sum()
    plt.pie(df_columns,labels=df_columns,autopct='%1.1f%%', startangle=140)
    plt.legend(df.columns)
    plt.show()
def donut():
    print ("\nDonut chart to show comparision of columns")
    df_columns=df.select_dtypes(include='number').sum()
    plt.pie(df_columns,labels=df_columns,autopct='%1.1f%%', startangle=140)
    plt.legend(df.columns)
    plt.gca().add_artist(plt.Circle((0,0),0.4,color="white"))
    plt.show()
def boxplot():
    print ("Boxplots for comparision of data before and after removing outliers")
    print (df.columns)
    ip1=input("\nEnter column name for x axis(Please enter correct spelling):")
    ip1=ip1.upper()

    mean=df[ip1].mean()
    std=df[ip1].std()
    ul=mean+1.5*std
    ll=mean-1.5*std
    fil=df[(df[ip1]<=ll)|(df[ip1]>=ul)]

    plt.figure(figsize=(10,4))
    plt.subplot(1,2,1)
    plt.boxplot(df[ip1])
    plt.xlabel(ip1)
    plt.title("Before Removing Outliers")
    plt.subplot(1,2,2)
    plt.boxplot(fil[ip1])
    plt.title("After Removing Outliers")
    plt.xlabel(ip1)
    plt.show()

while True:
    opt=int(input("\n\n1)Data Information \n2)Data Visualization \n3)Exit \nSelect your choice: "))
    if opt==1:
        while True:
            sel=int(input("\n1)Display empty columns \n2)Display number of rows and columns \n3)Display datatypes of columns \n4)Display column names \n5)Display top 10 rows \n6)Display last 10 rows \n7)Mean \n8)Median \n9)Mode \n10)Standard Deviation \n11)Variance \n12)Descriptive Statistics \n13)Back to menu \nSelect your choice: "))
            if sel==1:
                print ("\nEmpty column list:"),empty()
            elif sel==2:
                print ("\nNumber of rows and columns:"),shape()
            elif sel==3:
                print ("\nDatatypes of columns:"),dtypes()
            elif sel==4:
                print ("\nColumn names:"),columns()
            elif sel==5:
                print ("\nTop 10 rows:"),head()
            elif sel==6:
                print ("\nLast 10 rows:"),tail()
            elif sel==7:
                print ("\nMean of the Selected column is:"),mean()
            elif sel==8:
                print ("\nMedian of the Selected column is:"),median()
            elif sel==9:
                print ("\nMode of the Selected column is:"),mode()
            elif sel==10:
                print ("\nStandard Deviation of the Selected column is:"),std()
            elif sel==11:
                print ("\nVariance of the Selected column is:"),var()
            elif sel==12:
                print ("\nDescriptive Statistics of data:"),des()
            elif sel==13:
                print ("\nThank you")
                break
            else:
                print ("\nInvalid Choice, Select from menu options")
                continue
    elif opt==2:
        while True:
            sel=int(input("\n1)Histogram \n2)Bar Graph \n3)Scatter Plot \n4)Line Chart \n5)Heatmap \n6)Pie Chart \n7)Donut Chart \n8)Boxplot \n9)Back to menu \nSelect your choice: "))
            if sel==1:
                hist()
            elif sel==2:
                bar()
            elif sel==3:
                scatter()
            elif sel==4:
                line()
            elif sel==5:
                heatmap()
            elif sel==6:
                pie()
            elif sel==7:
                donut()
            elif sel==8:
                boxplot()
            elif sel==9:
                print ("\nThank you")
                break
            else:
                print ("\nInvalid choice, Select from menu options")
                continue
    elif opt==3:
        print("\nThank You")
        break
    else:
        print("\nInvalid Choice, Select from menu options")
        continue

```
