## Data Cleaning 

```python
data.isnull().sum()
```

```python
data.dropna(inplace=True)
```

```python
data.duplicated().sum()
```

```python
data[data.duplicated()]
```

```python
data['host_id']=data['host_id'].astype(object)
```

```python
data.info()
```

```python
sns.boxplot(data=data,x='price')
```

```python
df = data[data['price']<1500]
```

```python
sns.boxplot(data=df,x='price')
```

## Data Analysis and Visualization (EDA)
### Univariate Analysis

**Price Distribution**

```python
plt.figure(figsize=(8,5))
sns.histplot(data=df,x='price',bins=100)
plt.ylabel('Frequency')
plt.xlabel('Price')
plt.title('Price distribution')
plt.show()
```

*Insight - Most of the rooms are priced between $ 10 to 300* 

```python
df.columns
```

**Availability Distribution**

```python
plt.figure(figsize=(8,5))
sns.histplot(data=df,x='availability_365')
plt.ylabel('Frequency')
plt.xlabel('Days')
plt.title('Availability_365 distribution')
plt.show()
```

*Insight - Most of the rooms are available in the beginning and end of the year* 

**Average price per room based on neighbourhood_group**

```python
df.groupby(by='neighbourhood_group')['price'].mean()
```

*Insight - Rooms in Manhattan are expensive and rooms in Bronx are the cheapest* 

**Feature Engineering**

**Average price per bed based on neighbourhood_group**

```python
df['price per bed']=df['price']/df['beds']
df.head()
```

```python
df.groupby(by='neighbourhood_group')['price per bed'].mean()
```

*Insight - Beds in Manhattan are expensive and beds in Staten Island are the cheapest* 

## Bivariate Analysis
One variable dependency on another variable

```python
df.columns
```

**Comparison of price variance in different neighbour groups based on room type**

```python
sns.barplot(data=df, x='neighbourhood_group',y='price',hue='room_type')
plt.title('Price dependency on neigbourhood group')
```

*Insight :-* 
1) *Brooklyn has Expensive Homes or Appartments but Cheaper Private Rooms*
2) *Manhattan has Expensive Hotel Rooms but Cheaper Shared Rooms*
3) *Queens has Expensive Homes or Appartments but Cheaper Shared Rooms*
4) *Bronx has Expensive Homes or Appartments bu Cheaper Private Rooms*
5) *Staten Island has Expensive Shared Rooms but Cheaper Private Rooms*
6) *We can say, on an average Private Rooms are Cheaper in all neighbourhood groups*

**Comparison of Price dependency on reviews based on neighbourhood_group**

```python
plt.figure(figsize=(8,5))
sns.scatterplot(data=df,x='number_of_reviews',y='price',hue='neighbourhood_group')
plt.title ('Price dependency on reviews')
```

*Insight - There is negative correlation between Price and Number of Reviews* 

**Pairplots for Comparision**

```python
sns.pairplot(data=df,vars=['price','minimum_nights','number_of_reviews','availability_365'],hue='room_type')
```

*Insight - Shows the Correlation between the selected Columns* 

**Geographical Distribution Based on Room Type**

```python
plt.figure(figsize=(10,8))
sns.scatterplot(data=df,x='longitude',y='latitude',hue='room_type')
plt.title('Geographical Distribution of Listing ')
```

```python
df.dtypes
```

**Correlation Between columns**

```python
corr=df[['latitude', 'longitude', 'price', 'minimum_nights', 'number_of_reviews', 'reviews_per_month', 'availability_365', 'beds']].corr()
corr
```

*Insight - Shows the correlation between every column* 

**Correlation Heatmap Between columns**

```python
plt.figure(figsize=(8,4))
sns.heatmap(data=corr, annot=True)
plt.title('Correlation Between columns')
```

*Insight - Shows the correlation heatmap between every column*
